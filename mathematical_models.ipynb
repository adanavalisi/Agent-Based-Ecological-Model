```python
import numpy as np
import matplotlib.pyplot as plt
import math

# Parameters
alpha_0 = 0.1  # Initial value of alpha
R = 8.314  # Gas constant
T0 = 15  # Initial temperature
Ea = 50  # Activation energy
D = 1  # Dilution rate
m1, m2 = 2.11, 1.65  # Maximum growth rates
y1, y2 = 2 * m1, 2 * m2 / 3  # Yield factors
a1, a2 = 0.9, 2  # Half-saturation constants
D1, D2 = 2.018, 1.5  # Death rates
b = 1.8  # Amplitude of the sinusoidal term
w = 0.2  # Angular frequency of the sinusoidal term
a = 4
L0 = 0.2
I0 = 7
K0 = 0.005

# Initial conditions
S0 = 50  # Initial concentration of grass
x1_0, x2_0 = 200, 200  # Initial populations of species A and B

# Simulation parameters
t_max = 2000
dt = 0.1

# Arrays to store simulation results
t_values = np.arange(0, t_max, dt)
S_values = np.zeros_like(t_values)
x1_values = np.zeros_like(t_values)
x2_values = np.zeros_like(t_values)

# Initialize variables
S = S0
x1 = x1_0
x2 = x2_0

# Simulation loop
for i, t in enumerate(t_values):
    K = K0 + 0.0000001 * t
    T = T0 + 0.1 * t
    L = L0 - 0.05 * t
    #T = math.sin(t)

    # Calculate alpha(T)
    alpha_t = alpha_0 * np.exp(-R * T) / K

    # Calculate M with the sinusoidal term
    M = a * I0 * math.exp(-K * L)
    # Resource dynamics
    dS = (S0 + (M * math.sin(w * t))) * D - ((m1 / y1) * (x1 * S / (a1 + S))) - ((m2 / y2) * (x2 * S / (a2 + S)))
    S = max(S + dS * dt, 0)

    # Population dynamics
    dx1 = (m1 * x1 * S / (a1 + S)) - D1 * x1
    dx2 = (m2 * x2 * S / (a2 + S)) - D2 * x2
    x1 += dx1 * dt
    x2 += dx2 * dt

    # Store values
    S_values[i] = S
    x1_values[i] = x1
    x2_values[i] = x2

# Plot results
plt.figure(figsize=(12, 8))
plt.plot(t_values, S_values, label='Grass (S)')
plt.plot(t_values, x1_values, label='Species A (x1)')
plt.plot(t_values, x2_values, label='Species B (x2)')
plt.xlabel('Time')
plt.ylabel('Population/Concentration')
plt.legend()
plt.title('Species and Population Dynamics Simulation')
plt.show()
```
